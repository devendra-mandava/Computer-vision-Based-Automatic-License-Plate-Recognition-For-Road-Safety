```python
import os
import cv2
import shutil

# Input and output directories
source_dir = "/content/drive/MyDrive/ColabNotebooks/ALPR/dataset/green"
output_dir = "/content/drive/MyDrive/ColabNotebooks/ALPR/dataset/yolo1"
os.makedirs(output_dir, exist_ok=True)

# Parse the filenames and validate
def parse_filename(filename, img_width, img_height):
    try:
        # Split by underscores
        parts = filename.split('_')

        # Validate parts length
        if len(parts) < 5:
            raise ValueError(f"Invalid filename format: {filename}")

        # Extract class ID (ensure it's numeric)
        class_id = int(parts[0]) if parts[0].isdigit() else 0  # Default to class_id = 0 for unexpected values

        # Extract bounding box coordinates (assuming format x1_y1_x2_y2)
        x1, y1, x2, y2 = map(float, parts[1:5])

        # Ensure bounding box coordinates are valid
        if not (0 <= x1 < x2 <= img_width and 0 <= y1 < y2 <= img_height):
            raise ValueError(f"Invalid bounding box coordinates: {filename}")

        # Normalize coordinates for YOLO format
        x_center = ((x1 + x2) / 2) / img_width
        y_center = ((y1 + y2) / 2) / img_height
        width = (x2 - x1) / img_width
        height = (y2 - y1) / img_height

        return class_id, x_center, y_center, width, height

    except Exception as e:
        print(f"Skipping invalid file: {filename} -> {e}")
        return None  # Return None if the filename cannot be parsed

# Organize dataset into train/val splits
images = [f for f in os.listdir(source_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
train_split = int(0.8 * len(images))

# Split into train and val
train_images = images[:train_split]
val_images = images[train_split:]

for split, split_images in zip(['train', 'val'], [train_images, val_images]):
    split_img_dir = os.path.join(output_dir, split, 'images')
    split_lbl_dir = os.path.join(output_dir, split, 'labels')
    os.makedirs(split_img_dir, exist_ok=True)
    os.makedirs(split_lbl_dir, exist_ok=True)

    for img in split_images:
        # Read image to get its dimensions
        img_path = os.path.join(source_dir, img)
        image = cv2.imread(img_path)

        # Ensure the image is read properly
        if image is None:
            print(f"Skipping unreadable image: {img}")
            continue

        img_height, img_width = image.shape[:2]  # Get height and width dynamically

        # Move image
        shutil.copy2(img_path, os.path.join(split_img_dir, img))

        # Create corresponding label
        label_file = os.path.splitext(img)[0] + '.txt'
        result = parse_filename(img, img_width, img_height)
        if result:  # Only write labels if parsing is successful
            class_id, x_center, y_center, width, height = result
            with open(os.path.join(split_lbl_dir, label_file), 'w') as f:
                f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```

```python
import os
import xml.etree.ElementTree as ET
import shutil

# Input and output directories
source_dir = "/content/dataset/dataset"
output_dir = "/content/dataset/yolo"
os.makedirs(output_dir, exist_ok=True)

# Function to convert VOC (XML) to YOLO format
def convert_voc_to_yolo(xml_file, img_width, img_height):
    yolo_labels = []
    try:
        tree = ET.parse(xml_file)
        root = tree.getroot()

        for obj in root.findall('object'):
            class_id = 0  # Since it's a single-class problem
            bndbox = obj.find('bndbox')
            x1 = float(bndbox.find('xmin').text)
            y1 = float(bndbox.find('ymin').text)
            x2 = float(bndbox.find('xmax').text)
            y2 = float(bndbox.find('ymax').text)

            # Normalize YOLO format
            x_center = ((x1 + x2) / 2) / img_width
            y_center = ((y1 + y2) / 2) / img_height
            width = (x2 - x1) / img_width
            height = (y2 - y1) / img_height

            yolo_labels.append(f"{class_id} {x_center} {y_center} {width} {height}")

        return yolo_labels
    except Exception as e:
        print(f"Error converting {xml_file}: {e}")
        return None

# Process train and val sets
for split in ['train', 'test']:
    split_src_dir = os.path.join(source_dir, split)
    split_img_dir = os.path.join(output_dir, 'train/images' if split == 'train' else 'val/images')
    split_lbl_dir = os.path.join(output_dir, 'train/labels' if split == 'train' else 'val/labels')
    os.makedirs(split_img_dir, exist_ok=True)
    os.makedirs(split_lbl_dir, exist_ok=True)

    for file in os.listdir(split_src_dir):
        if file.endswith('.jpg'):
            # Copy image to the output directory
            img_path = os.path.join(split_src_dir, file)
            shutil.copy2(img_path, os.path.join(split_img_dir, file))

            # Find corresponding XML file and convert to YOLO format
            xml_file = os.path.splitext(img_path)[0] + '.xml'
            if os.path.exists(xml_file):
                # Read image dimensions
                import cv2
                image = cv2.imread(img_path)
                img_height, img_width = image.shape[:2]

                # Convert XML to YOLO and save
                yolo_labels = convert_voc_to_yolo(xml_file, img_width, img_height)
                if yolo_labels:
                    label_file = os.path.join(split_lbl_dir, os.path.splitext(file)[0] + '.txt')
                    with open(label_file, 'w') as f:
                        f.write('\n'.join(yolo_labels))
            else:
                print(f"XML file not found for image: {file}")

print("Dataset conversion complete!")
```
